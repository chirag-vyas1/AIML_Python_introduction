## <a id ="sub1-1" ><a/>**`1. Console Input`**

- Console Input/Output means input from keyboard and output to screen.
- Console input can be received using the built-in **`input()`** function.

- **General form** of `input()` function is:
```python
s = input('prompt')
```
- prompt is a string that is displayed on the screen, soliciting a value.**input()** returns a string.**Example:** If 123 is entered as input, '123' is returned.
- **input()** can be used to receive 1 or more values.
  
```python
# Receive full name
name = input('Enter full name')

# Separate first name, middle name and surname
fname, mname, lname = input('Enter full name: ').split()

```
- **`split()`**function will split the entered fullname with space as delimiter.
The split values will then be assigned to fname, mname, lname.
- If we are to receive multiple int values, we can receive them as strings and then convert them to ints.

```python
n1, n2, n3 = input('Enter three values: ').split()
n1, n2, n3 = int(n1), int(n2), int(n3)
print(n1 + 10, n2 + 20, n3 + 30)
```
- The same thing can be done in a more compact manner using list comprehension. it applies int() function to element of the list returned by the split() function.

```python
n1, n2, n3 = [int(n) for n in input('Enter three values: ').split()]
print(n1 + 10, n2 + 20, n3 + 30)
```

- The expression enclosed within `[ ]` is called list comprehension. It will be discussed in detail in Chapter 12.
  **`input()`** can be used to receive an arbitrary number of values.

## <a id ="sub1-2" ><a/>**`2. Console Output`**
- Built-in function print() is used to send output to screen.
- print() function has this form:

```python
print(objects, sep=' ', end='\n', file=sys.stdout, flush=False)
```

- This means that by default:
- Objects are printed to screen (sys.stdout)
- Separated by space (sep=' ')
- End with a newline (end='\n')
- flush=False means the output buffer will not be flushed immediately
- Python allows calling functions and passing keyword arguments.
So, while calling print(), we can pass custom values for sep and end:

```python
print(a, b, c, sep=',', end='!')    # prints a,b,c!
print(x, y, sep='...', end='#')     # prints x...y#
```

### <a id ="sub1-3" ><a/>**`3. Formatted Printing`**
- There are 4 ways to control output formatting:
  
- (a) Using formatted string literals - easiest
- (b) Using the format() method - older
- (c) C-style printf() formatting - legacy
- (d) Slicing and concatenation - difficult

Today, (a) is most commonly used, followed by (b).

- using formatted string literals (often called fstrings):

```python
r, l, b = 1.5678, 10.5, 12.66
print(f'radius = {r}')
print(f'length = {l} breadth = {b} radius = {r}')

name = 'Sushant Ajay Raje'
for n in name.split():
    print(f'{n:10}')  # print in 10 columns
```
### Using **`format()`** method of string object.

```python
r, l, b = 1.5678, 10.5, 12.66
name, age, salary = 'Rakshita', 30, 53000.55

# Print in order by position of variables using empty {}
print('radius = {} length = {} breadth = {}'.format(r, l, b))
print('name = {} age = {} salary = {}'.format(name, age, salary))

# Print in any desired order
print('breadth = {2} length = {1} radius = {0}'.format(r, l, b))
print('age = {1} salary = {2} name = {0}'.format(name, age, salary))

# Print name in 15 columns, salary in 10 columns
print('name = {:15} salary = {:10}'.format(name, salary))

# Print radius in 10 columns, with 2 digits after decimal point
print('radius = {:10.2f}'.format(r))

### `Problem 7.1`
Write a program to receive radius of a circle, and length and breadth of a rectangle in one call to input( ). Calculate and print the circumference of circle and perimeter of rectangle.

### **`Program`**


````python
r, l, b = input('Enter radius, length and breadth: ').split()
radius = float(r)
length = float(l)
breadth = float(b)
circumference = 2 * 3.14 * radius
perimeter = 2 * (length + breadth)
print(circumference)
print(perimeter)
````

### **`Tips`**
- **input( )** returns a string, so it is necessary to convert it into int or float suitably, before performing arithmetic operations.

## **`Problem 7.2`**
Write a program to receive 3 integers using one call to input( ). The three integers signify starting value, ending value and step value for a range. The program should use these values to print the number, its square and its cube, all properly right-aligned. Also output the same values left-aligned.

````python
start, end, step = input('Enter start, end, step values: ').split()
# right aligned printing
for n in range(int(start), int(end), int(step)):
    print(f'{n:>5}{n**2:>7}{n**3:>8}')
print()

# left aligned printing
for n in range(int(start), int(end), int(step)):
    print(f'{n:<5}{n**2:<7}{n**3:<8}')
````

## **`Tips`**
- **`
{n:>5`**} will print n right-justified within 5 columns. Use < to left-justify
- **`{0:<5}`** will left-justify 0th parameter in the list within 5 columns. Use > to right-justi

Suppose there are 5 variables in a programmax, min, mean, sd and var, having some suitable values. Write a program to print these variables properly aligned using multiple fstrings, but one call to print()

## **`Problem 7.5`**
Write a program that prints square root and cube root of numbers from 1 to 10, up to 3 decimal places. Ensure that the output is displayed in separate lines, with number center-justified and square and cube root, right-justified
## **`Program`**.

````python
# Problem 7.5 Solution
import math

print(f"{'Number':^10}{'Square Root':>12}{'Cube Root':>12}")
print("-" * 34)

for n in range(1, 11):
    s = round(math.sqrt(n), 3)
    c = round(math.pow(n, 1/3), 3)
    print(f"{n:^10}{s:>12.3f}{c:>12.3f}")
````
