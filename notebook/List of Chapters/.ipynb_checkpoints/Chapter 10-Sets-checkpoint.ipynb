## <a id ="sub1-1"><a/>**`1. What are Sets?`**

- Sets are like lists, with an exception that they do not contain duplicate entries.

```python
a = set()    # empty set, use set() instead of {}
b = {20}     # set with one item
c = {'Sanjay', 25, 34555.50}    # set with multiple items
d = {10, 10, 10, 10}    # only one 10 gets stored

```
- While storing an element in a set, its hash value is computed using a hashing technique to determine where it should be stored in the set.
- Since hash value of an element will always be same, no matter in which order we insert the elements in a set, they get stored in the same order.

```python
s = {12, 23, 45, 16, 52}
t = {16, 52, 12, 23, 45}
u = {52, 12, 16, 45, 23}

print(s)    # prints {12, 23, 45, 16, 52}
print(t)    # prints {12, 23, 45, 16, 52}
print(u)    # prints {12, 23, 45, 16, 52}
```

- It is possible to create a set of strings and tuples, but not a set of lists.

```python
s1 = {'Morning', 'Evening'}    # works
s2 = {(12, 23), (15, 25), (17, 34)}    # works
s3 = {[12, 23], [15, 25], [17, 34]}    # error
```
- Since strings and tuples are immutable, their hash value remains same at all times. Hence a set of strings or tuples is permitted. However, a list may change, so its hash value may change, hence a set of lists is not permitted.
- Sets are commonly used for eliminating duplicate entries and membership testing.

## <a id ="sub1-2"><a/>**`2. Accessing Set Elements`**

- Entire set can be printed by just using the name of the set. Set is an unordered collection. Hence order of insertion is not same as the order of access.

```python
s = {15, 25, 35, 45, 55}
print(s)    # prints {35, 45, 15, 55, 25}
```
- Being an unordered collection, items in a set cannot be accessed using indices.
- Sets cannot be sliced using `[ ]`.

## <a id ="sub1-4"><a/>**`4. Basic Set Operations`**
- Sets like lists are mutable. Their contents can be changed.

```python
s = {'gate', 'fate', 'late'}
s.add('rate')    # adds one more element to set 
```
- If we want an immutable set, we should use a frozenset.

```python
s = frozenset({'gate', 'fate', 'late'})
s.add('rate')    # error
```
- Given below are the operations that work on lists and tuples. These operations are discussed in detail in Chapter 8. Try these operations on sets too.

**`Set Operations Summary`**

**Operations that don't work with sets:**
- Concatenation - doesn't work
- Merging - doesn't work

**Operations that work with sets:**
- Conversion - works
- Aliasing - works
- Cloning - works
- Searching - works
- Identity - works
- Comparison - works
- Emptiness - works

- Two sets cannot be concatenated using `+`.
- Two sets cannot be merged using the form `z = s + t`.
- While converting a set using `set()`, repetitions are eliminated.

```python
list = [10, 20, 10, 30, 40, 50, 30]
s = set(list)    # will create set containing {10, 20, 30, 40, 50}  s

## <a id ="sub1-6"><a/>**`6.Set Methods`**
- Any set is an object of type **set**. Its methods can be accessed using the syntax **s.method()**.usage of commonly used set method is shown below.

```python

s = {12, 15, 13, 23, 22, 16, 17}
t = {'A', 'B', 'C'}
u = set()     # empty set

s.add('Hello')    # adds 'Hello' to s
s.update(t)       # adds elements of t
```
**`Set Methods (Continued)`**

```python
u = s.copy()        # performs deep copy (cloning)
s.remove(15)        # deletes 15 from s
s.remove(101)       # would raise error, as 101 is not a member of s
s.discard(12)       # removes 12 from s
s.discard(101)      # won't raise an error, though 101 is not in s
s.clear()           # removes all elements
```
**`Set Relationships`**
- Following methods can be used on 2 sets to check the relationship between them:

```python
s = {12, 15, 13, 23, 22, 16, 17}
t = {13, 15, 22}

print(s.issuperset(t))    # prints True
print(s.issubset(t))      # prints False
print(s.isdisjoint(t))    # prints False
```
Since all elements of **t** are present in **s**:
- **s** is a superset of **t**
- **t** is subset of **s**
- If intersection of two sets is null, the sets are called disjoint sets to s

```
- It is possible to unpack a set using the * operator:

```python

### **`[B] Set Operations - Questions and Answers`**

**(a) Separate Names Beginning with A and B:**
- **Question:** A set contains names which begin either with A or with B. Write a program to separate out the names into two sets.

````python
names = {'Alice', 'Bob', 'Anna', 'Brian', 'Amy', 'Brad', 'Charlie'}
a_names = {name for name in names if name.startswith('A')}
b_names = {name for name in names if name.startswith('B')}

print("Names starting with A:", a_names)
print("Names starting with B:", b_names)
````

**(b) Empty Set Operations**
- **Question:** Create an empty set. Write a program that adds five new names, modifies one existing name and deletes two names.

````python
# Create empty set
names_set = set()

# Add five names
names_set.update(['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'])
print("After adding:", names_set)

# Modify one name (remove and add)
names_set.remove('Bob')
names_set.add('Robert')
print("After modifying:", names_set)

# Delete two names
names_set.remove('Alice')
names_set.discard('Charlie')
print("After deletion:", names_set)
````

**(d) Write a program to create a set containing 10 randomly generated number in the range 15 to 45. count howmany of these numbers are less than 30.delete all numbers which are reater than 35**

````python
import random

# Create set with 10 random numbers
numbers = {random.randint(15, 45) for _ in range(10)}
print("Original set:", numbers)

# Count numbers less than 30
count_less_than_30 = len([x for x in numbers if x < 30])
print("Numbers less than 30:", count_less_than_30)

# Delete numbers greater than 35
numbers = {x for x in numbers if x <= 35}
print("After deleting numbers > 35:", numbers)
````

**(g) How will you remove all duplicate element present in a string, a list and a tuple?**

````python
# String
text = "programming"
unique_chars = ''.join(set(text))
print("String without duplicates:", unique_chars)

# List
my_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = list(set(my_list))
print("List without duplicates:", unique_list)

# Tuple
my_tuple = (1, 2, 2, 3, 4, 4, 5)
unique_tuple = tuple(set(my_tuple))
print("Tuple without duplicates:", unique_tuple)
````

**(h) Which operator is used for determining whether a set is a subset of another set?**

````python
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}
print(set1 <= set2)  # True
print(set1.issubset(set2))  # True
````
