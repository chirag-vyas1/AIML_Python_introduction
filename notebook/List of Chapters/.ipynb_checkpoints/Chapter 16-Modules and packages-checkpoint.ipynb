### <a id="sub16-1"><a/>**`1.The Main Module`**

- A module is a `.py` file containing definitions and statements. So, all `.py` files that we created so far for our programs are modules.

- When we execute a program, its module name is `__main__`. This name is available in the variable `__name__`.

```python
def display():
    print('You cannot make History if you use Incognito Mode')

def show():
    print('Pizza is a pie chart of how much pizza is left')

print(__name__)
display()
show()
```

**Output:**

```
__main__
You cannot make History if you use Incognito Mode  
Pizza is a pie chart of how much pizza is left
```

### <a id="sub16-2"><a/>**`2.Multiple Modules`**

- There are two reasons why we may want to create a program that contains multiple modules:

#### (a)
It makes sense to split a big program into multiple `.py` files, where each `.py` file acts as a module.

**Benefit**  Ease of development and maintenance.

#### (b)
We may need a set of handy functions in several programs.  
In such a case, instead of copying these functions in different program files, we may keep them in one file and use them in different programs.

**Benefit**  Reuse of existing code.


### <a id="sub16-3"><a/> **`3.Importing a Module`**

- To use the definitions and statements in a module in another module, we need to `import` it into this module.

#### Example 1:

**`functions.py`**

```python
def display():
    print('Earlier rich owned cars, while poor had horses')

def show():
    print('Now everyone has car, while only rich own horses')
```

**`usefunctions.py`**

```python
import functions

functions.display()
functions.show()
```

When we execute `usefunctions.py`, it runs as a module with name `__main__`.

- `import functions` makes the definitions in `functions.py` available in `usefunctions.py`.

---

- A module can import multiple modules.

```python
import math
import random
import functions  # use functions.py from the previous example

a = 100
b = 200

print(__name__)
print(math.sin(0.5))
print(math.cos(0.5))
print(random.random())
print(random.randint(30, 45))
functions.display()
functions.show()
```

Here, `__name__` contains `__main__`, indicating that we are executing the main module.  
`random` and `math` are standard modules.  
`functions` is a user-defined module.

## <a id="sub16-4"><a/> **`4. Variations of import`**

- The `import` statement can be used in multiple forms:

```python
import math
import random
```

is the same as:

```python
import math, random
```

---

- If we wish, we can import specific names from a module:

```python
from math import sin, cos, tan
from functions import display      # imports only display function
from functions import *            # imports all functions
```

---

- We can rename a module while importing it. We can then use the new name in place of the original module name:

```python
import functions as fun
fun.display()
```

or even:

```python
from functions import display as disp
disp()
```

---



## <a id = "sub16-5"><a/> **`5.Search Sequence`**
`
- If we import a module called `'myfuncs'`, the following search sequence will be followed:

  - The interpreter will first search for a built-in module called `'myfuncs'`.
  - If such a module is not found, then it will search for it in the directory list given by the variable `sys.path`.

---

- The list in the `sys.path` variable contains:

  - The directory from where the script has been executed,
  - Followed by a list of directories as specified in the `PYTHONPATH` environment variable.

---

- We can print the list of directories in `sys.path` using:

```python
import sys
for p in sys.path:
   print(p)
```

````python
import sys
for p in sys.path:
   print(p)
````

### <a id ="sub=16-6"><a/> **`6.Same Code, Different Interpretation`**

- Suppose we have a module called `functions` in `'functions.py'`. If this module has functions `display()` and `main()`, we want to use this program sometimes as an independent script, and at other times as a module (from which we can use `display()` function).

- To achieve this, we write the code in this fashion:

```python
# functions.py
def display():
    print("Wright Brothers are responsible for 9/11 too")

def main():
    print("If you beat your own record, you win as well as lose")
    print("Internet connects people at a long distance")
    print("Internet disconnects people at a short distance")
    display()

if __name__ == '__main__':
    main()
```

- If we run it as an independent program, the condition will be satisfied. As a result, `main()` will be called.
- If we import this module in another program, the condition will fail, so `main()` will not be called. However, the program can still call `display()` independently.

### <a id ="sub=16-7"><a/>**`7. Packages`**

- The way drives, folders, and subfolders help us organize files in an OS, **packages help us organize sub-packages and modules**.

-  A particular directory is treated as a **package** if it contains a file named `__init__.py`.  
   The directory may contain:
  - Other sub-packages
  - Modules
  - The `__init__.py` file may be empty or contain initialization code.
    
    
### **`Packages Example`**

- Suppose there is a package called `pkg` containing a module called `mod.py`.
- If the module contains functions `f1()` and `f2()`, then the directory structure would be as follows:

```
Directory - pkg
     mod.py
     __init__.py
```

- **Contents of `mod.py`**  functions `f1()` and `f2()`:

```python
# mod.py
def f1():
    print('Inside function f1')

def f2():
    print('Inside function f2')
```

- **Program to use `f1()` and `f2()` (client.py):**

```python
# client.py
import pkg.mod

pkg.mod.f1()
pkg.mod.f2()
```

### <a id="sub16-8"><a/>**`8. Third-party Packages`**

- Pythonistas in the Python community create software and make it available for other programmers to use.
- They use **PyPI**  Python Package Index ([https://www.pypi.org](https://www.pypi.org)) to distribute their software.
- PyPI maintains the list of third-party Python packages available.

#### Key Points:

- There are third-party packages available for literally everything under the sun.
- You too can register at **PyPI** and upload your packages.
- Guidelines for publishing: [https://www.pypi.org](https://www.pypi.org)

---

### Installing Packages with `pip`

- To use a package available on PyPI:
  1. Download it.
  2. Install it using a package manager like `pip`.

- `pip` is installed automatically with Python.

#### Example command to install a package (`pykrige`) from PyPI:

```bash
c:\> pip install pykrige
```
