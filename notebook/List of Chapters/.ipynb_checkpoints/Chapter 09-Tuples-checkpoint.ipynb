## **`1.What are Tuples?`**

- A list can store dissimilar data, but it is commonly used for storing similar data.
- A tuple, though capable of storing similar data, is commonly used for storing dissimilar data.
- Tuple data is enclosed within parentheses `( )`.

**Examples of Tuple Creation**

```python
a = ()                         # Empty tuple
b = (10,)                      # Tuple with one item (note the comma)
c = ('Sanjay', 25, 34555.50)   # Tuple with dissimilar items
d = (10, 20, 30, 40)           # Tuple with similar items

```
- **Note:** If we omit the comma in b = (10), it will be treated as an integer, not a tuple.
- While initializing a tuple, we may drop()
  
```python
c = 'Sanjay', 25, 34555.50     # Tuple with multiple items
print(type(c))                 # Output: <class 'tuple'>
```
- Like lists, elements in a tuple can be repeated using the multiplication operator *.
```python
tpl1 = (10,) * 5               # Stores (10, 10, 10, 10, 10)
tpl2 = (10) * 5                # Stores 50 (not a tuple)


## **`2.Accessing Tuple Elements`**
- The entire tuple can be printed using its name.

```python
tpl = ('Sanjay', 25, 34555.50)
print(tpl)

```
- Tuples are ordered collections.
- Elements can be accessed by index (starting from 0), just like strings and lists.

```python
msg = ('Handle', 'Exceptions', 'Like', 'a', 'boss')
print(msg[1], msg[3])          # Outputs: Exceptions a
```

**Tuple Slicing**

- Like strings and lists, tuples can also be sliced to get smaller tuples.

```python
emp = ('Sanjay', 23, 23000, 1760, 2040)
print(emp[1:3])     # Prints (23, 23000)
print(emp[3:])      # Prints (1760, 2040)
print(emp[:3])      # Prints ('Sanjay', 23, 23000)

## **`3. Looping in Tuples`**
- We can iterate through tuples using while or for loops.

```python
tpl = (10, 20, 30, 40, 50)
i = 0
while i < len(tpl):
    print(tpl[i])
    i += 1

### Using a for loop:

```python
for n in tpl:
    print(n)
    
- To keep track of the index while looping, use **enumerate():**

```python
tpl = (10, 20, 30, 40, 50)
for index, n in enumerate(tpl):
    print(index, n)

## **`4. Basic Tuple Operations`**
- **Mutability:** Tuples are immutable, meaning their content cannot be modified.
```python
msg = ('Fall', 'In', 'Line')
msg[0] = 'FALL'                    # Error
msg[1:3] = ('Above', 'Mark')      # Error

- Operations like append, remove, and insert are not allowed on tuples.
- Although a tuple is immutable, it can contain mutable objects like lists.
```
**Mutable Objects in Tuples**

- Tuples can contain both mutable and immutable objects.

```python
s = ([1, 2, 3, 4], [4, 5], 'Ocelot')
```
- If a tuple contains a list, the list can be modified, because lists are mutable.
```python
s = ([1, 2, 3, 4], [10, 20], 'Onyx')
s[1][1] = 45
print(s)       # prints ([1, 2, 3, 4], [10, 45], 'Onyx')
```
- Another way to modify a mutable item within a tuple:

```python
p = s[1]
p[1] = 100
print(s)       # prints ([1, 2, 3, 4], [10, 100], 'Onyx')
```
**Common Tuple Operations**
These are operations similar to list operations that can also be tried on tuples:

- Concatenation
- Merging

- Aliasing
- Cloning

- Identity
- Comparison
- Emptiness


## **`5. Using Built-in Functions on Tuples`**
- Several built-in functions can be used directly on tuples:

```python

t = (12, 15, 13, 23, 22, 16, 17)

len(t)        # returns number of elements
max(t)        # returns maximum value
min(t)        # returns minimum value
sum(t)        # returns the sum of all elements
any(t)        # returns True if any element is True
all(t)        # returns True if all elements are True
sorted(t)     # returns a new sorted list (not a tuple)
reversed(t)   # returns a reverse iterator (needs to be converted using list())


## **`6.Tuple Methods`**

Any tuple is an object of type `tuple`. Its methods can be accessed using the syntax `tpl.method()`. Usage of two methods is shown below:

```python
tpl = (12, 15, 13, 23, 22)  # create tuple
print(tpl.count(23))        # return no. of times 23 appears in tpl
print(tpl.index(22))        # return index of item 22
print(tpl.index(50))        # reports ValueError as 50 is absent in tpl

- if we wish to sort a list of tuples or tuple of lists, it can be done as follows:

```python
import operator

lst = [('Shailesh', 24, 3455.50), ('Priyanka', 25, 4555.50)]  # List of tupple 
tpl = (['Shailesh', 24, 3455.50], ['Priyanka', 25, 4555.50])   # Tuple of sets 

# Default sorting (by first item, e.g., name)
print(sorted(lst))  
print(sorted(tpl))  

# Sort by salary (3rd item, index 2)
print(sorted(lst, key=operator.itemgetter(2)))  
print(sorted(tpl, key=operator.itemgetter(2)))  
```
- Default Behavior:**sorted()** sorts by the first item in each tupl /list (e.g., name in this case

- if we wish to sort by salary , we need to use the **itemgetter()** function of **operator** module.

- The **Key** parameter of **sorted()** requires a key function( tobe applied to objects to be sorted) rather then a single key value.

- **operator.itemgettr(2)** will give us a function that fetchs salary from a list / tuple.

- In general, **operator.itemgetter()** constructs function that takes a list/tuple as input, and fetches the nth element out of it.

## **`Problem 8.3`**  

A list contains names of boys and girls as its elements. Boys' names are stored as tuples. Write a Python program to find out number of boys and girls in the list. 

## **`Program`**

````python
lst = ['Shubha', 'Nisha', 'Sudha', ('Suresh',), ('Rajesh',), 'Radha']
boys =0
girls =0
lst = ['Shubha', 'Nisha', 'Sudha', ('Suresh',), ('Rajesh',), 'Radha']
boys = 0
girls = 0
for ele in lst:
    if isinstance(ele, tuple):
        boys += 1
    else:
        girls += 1
print('Boys =', boys, 'Girls =', girls)
````

## **`Tips`**
 isinstance( ) functions checks whether ele is an instance of tuple type.  
 Note that since the tuples contain a single element, it is followed by a comma.

## **`Problem 8.5`**

Given the following tuple  
`('F', 'I', 'a', 'b', 'b', 'e', 'r', 'g', 'a', 's', 't', 'e', 'd')`
Write a Python program to carry out the following operations:
- Add an ! at the end of the tuple
- Convert a tuple to a string
- Extract ('b', 'b') from the tuple
- Find out number of occurrences of 'e' in the tuple
- Check whether 'r' exists in the tuple
- Convert the tuple to a list
- Delete characters 'b', 'b', 'e', 'r' from the tuple

## **`Program`**

````python
tpl = ('F', 'I', 'a', 'b', 'b', 'e', 'r', 'g', 'a', 's', 't', 'e', 'd')

# addition of ! is not possible as tuple is an immutable
# so to add ! we need to create a new tuple and then make tpl refer to it
tpl = tpl + ('!',)
print(tpl)


# convert tuple to string
s = ''.join(tpl)
print(s)


# extract ('b', 'b') from the tuple
t = tpl[3:5]
print(t)


# count number of 'e' in the tuple
count = tpl.count('e')
print('count = ', count)


# check whether 'r' exists in the tuple
print('r' in tpl)


# Convert the tuple to a list

lst1 = list(tpl)
print(lst1)


# tuples are immutable, so we cannot remove elements from it
# we need to split the tuple, eliminate the unwanted element and then merge the tuples
tpl = tpl[:3] + tpl[7:]
print(tpl)
````

### (c) suppose a date is represented as a tuple (d,m,y), Write a program to create two date tuples and find the number of days between the two dates.

````python
from datetime import date

date1 = (28, 9, 2025)
date2 = (15, 10, 2025)

d1 = date(date1[2], date1[1], date1[0])
d2 = date(date2[2], date2[1], date2[0])

diff = abs((d2 - d1).days)
print("Days between:", diff)
````

### (d) Create a list of tuples. Each tuple should contain an item and its price in float.  
Write a program to sort the tuples in descending order by price.  
*Hint: Use `operator.itemgetter()`*

---




````python
from operator import itemgetter

items = [
    ("Pen", 10.5),
    ("Notebook", 45.0),
    ("Pencil", 5.0),
    ("Bag", 120.0),
    ("Eraser", 3.5)
]

sorted_items = sorted(items, key=itemgetter(1), reverse=True)
print("Sorted by price (descending):", sorted_items)
````

### (e) Store the data about shares held by a user as tuples containing the following information about shares:

- Share name  
- Date of purchase  
- Cost price  
- Number of shares  
- Selling price  

Write a program to determine:  
- Total cost of the portfolio  
- Total amount gained or lost  
- Percentage profit made or loss incurred  

---


````python
shares = [
    ("TCS", "01-01-2024", 3000, 10, 3500),
    ("Infosys", "15-02-2024", 1500, 20, 1400),
    ("Wipro", "10-03-2024", 400, 50, 500)
]

total_cost = 0
total_value = 0

for name, date, cost, qty, sell in shares:
    total_cost += cost * qty
    total_value += sell * qty

gain_loss = total_value - total_cost
percent = (gain_loss / total_cost) * 100

print("Total cost:", total_cost)
print("Total value:", total_value)
print("Gain/Loss:", gain_loss)
print("Percentage change:", percent, "%")
````

### (g) Write a program to create following 3 lists:  
- a list of names  
- a list of roll numbers  
- a list of marks  

Generate and print a list of tuples containing name, roll number and marks from the 3 lists. From this list generate 3 tuples:  
- one containing all names  
- another containing all roll numbers  
- and a third containing all marks.

````python
names = ["Alice", "Bob", "Charlie"]
rolls = [101, 102, 103]
marks = [85, 90, 95]

students = list(zip(names, rolls, marks))
print("List of tuples:", students)

names_tuple = tuple([s[0] for s in students])
rolls_tuple = tuple([s[1] for s in students])
marks_tuple = tuple([s[2] for s in students])

print("Names tuple:", names_tuple)
print("Roll numbers tuple:", rolls_tuple)
print("Marks tuple:", marks_tuple)
````
