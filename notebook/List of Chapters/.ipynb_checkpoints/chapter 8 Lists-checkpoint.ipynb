## <a id =sub1-3><a/> **`3 .Looping Through Lists`**
- Using Loops to Process List Items
- To process each item in a list, we can iterate through it using either a while loop or a for loop.

````python
animals = ['Zebra', 'Tiger', 'Lion', 'Jackal', 'Kangaroo']

# using while loop
i = 0
while i < len(animals):
    print(animals[i])
    i += 1

# using more convenient for loop
for a in animals:
    print(a)
````

- While iterating through a list using a for loop, if we wish to keep track of the index of the element that a is referring to, we can do so using the built-in enumerate() function.

````python
animals = ['Zebra', 'Tiger', 'Lion', 'Jackal', 'Kangaroo']
for index, a in enumerate(animals):
    print(index, a)
````

- Using sorted() and reversed() with Lists

## **`Queue Data Structure`**

- A queue is a First-In First-Out (FIFO) list.  
  The first element that is added to the list is the first element that is removed from it.

- **Why Python lists are not ideal for queues**

- Lists are not efficient for implementing queue data structures:
  - Removing items from the beginning (`list.pop(0)`) is inefficient because it shifts all remaining elements one position forward.
  - This operation becomes slow for large lists.

- **Using deque from collections module**

- The `deque` class from the `collections` module is preferred for queue implementation.
- `deque` stands for double-ended queue.
- It allows fast append and pop operations from both ends.

```python
from collections import deque

queue = deque()

# Enqueue operations
queue.append('A')
queue.append('B')
queue.append('C')

print(queue)  # deque(['A', 'B', 'C'])

# Dequeue operation
queue.popleft()       # 'A'
print(queue)          # deque(['B', 'C'])

## **`Problem 8.1`**
Perform the following operations on a list of names.

- Create a list of 5 names - 'Anil', 'Amol', 'Aditya', 'Avi', 'Alka'
- Insert a name 'Anuj' before 'Aditya'
- Append a name 'Zulu'
- Delete 'Avi' from the list
- Replace 'Anil' with 'Anilkumar'
- Sort all the names in the list
- Print reversed sorted list

## **`Program`**

````python
# Problem 8.1 Solution

# Create a list of 5 names
names = ['Anil', 'Amol', 'Aditya', 'Avi', 'Alka']
print("Original list:", names)

# Insert a name 'Anuj' before 'Aditya'
names.insert(2, 'Anuj')
print("After inserting 'Anuj':", names)

# Append a name 'Zulu'
names.append('Zulu')
print("After appending 'Zulu':", names)

# Delete 'Avi' from the list
names.remove('Avi')
print("After removing 'Avi':", names)

# Replace 'Anil' with 'Anilkumar'
i = names.index('Anil')
names[i] = 'Anilkumar'
print("After replacing 'Anil':", names)

# Sort all the names in the list
names.sort()
print("Sorted list:", names)

# Print reversed sorted list
names.reverse()
print("Reversed sorted list:", names)
````

## **`Problem 8.2`**
Perform the following operations on a list of numbers.

- Create a list of 5 odd numbers
- Create a list of 5 even numbers
- Combine the two lists
- Add prime numbers 11, 17, 29 at the beginning of the combined list
- Report how many elements are present in the list
- Replace last 3 numbers in the list with 100, 200, 300
- Delete all the numbers in the list
- Delete the list

## **`Program`**

````python
# Problem 8.2 Solution

# Create a list of 5 odd numbers
odd_numbers = [1, 3, 5, 7, 9]
print("Odd numbers:", odd_numbers)

# Create a list of 5 even numbers
even_numbers = [2, 4, 6, 8, 10]
print("Even numbers:", even_numbers)

# Combine the two lists
combined = odd_numbers + even_numbers
print("Combined list:", combined)

# Add prime numbers 11, 17, 29 at the beginning
combined = [11, 17, 29] + combined
print("After adding primes at beginning:", combined)

# Report how many elements are present
num_elements = len(combined)
print("Number of elements:", num_elements)

# Replace last 3 numbers with 100, 200, 300
combined[-3:] = [100, 200, 300]
print("After replacing last 3 elements:", combined)

# Delete all the numbers in the list
combined.clear()
print("After clearing the list:", combined)

# Delete the list
del combined
# print(combined)  # This would raise NameError as 'combined' no longer exists
````


## **`Problem 8.3`**
Write a program to implement a Stack data structure. Stack is a Last In First Out (LIFO) list in which addition and deletion takes place at the same end.
## **`Program`**

````python
# Problem 8.3 Solution - Stack Implementation (LIFO)

# Create an empty stack
stack = []
print("Initial stack:", stack)

# Push elements onto the stack (add to the end)
stack.append(10)
stack.append(20)
stack.append(30)
stack.append(40)
stack.append(50)
print("Stack after pushing elements:", stack)

# Pop elements from the stack (remove from the end - LIFO)
print("\nPopping elements (LIFO order):")
print("Popped:", stack.pop())
print("Popped:", stack.pop())
print("Popped:", stack.pop())

print("Stack after popping 3 elements:", stack)

# Demonstrate more operations
print("\n--- Additional Stack Operations ---")

# Check if stack is empty
print("Is stack empty?", len(stack) == 0)

# Peek at the top element without popping
if stack:
    print("Top element (peek):", stack[-1])

# Push more elements
stack.append(60)
stack.append(70)
print("After pushing 60 and 70:", stack)

# Pop all remaining elements
print("\nPopping all remaining elements:")
while stack:
    print("Popped:", stack.pop())

print("Final stack:", stack)
print("Is stack empty now?", len(stack) == 0)
````

## **`Problem 8.4`**
Write a program to implement a Queue data structure. Queue is a First In First Out (FIFO) list, in which addition takes place at the rear end of the queue and deletion takes place at the front end of the queue.
## **`Program`**

````python
# Problem 8.4 Solution - Queue Implementation (FIFO)
import collections

# Create an empty queue
queue = collections.deque()
print("Initial queue:", queue)

# Enqueue elements (add to the rear)
queue.append('Subana')
queue.append('Shabana')
queue.append('Shakila')
queue.append('Shakira')
queue.append('Sameera')
print("Queue after enqueuing elements:", queue)

print(queue.popleft())
print(queue.popleft())
print(queue.popleft())
````

````python
lst = ["apple", "banana", "cherry", "date"]
first_chars = [word[0] for word in lst]
print(first_chars)   # ['a', 'b', 'c', 'd']
````

**(j) A list contains 10 numbers. Write a program to eliminate all duplicates from the list.**

````python
lst = [1, 2, 2, 3, 4, 4, 5, 6, 7, 7]
lst = list(set(lst))
print(lst)   # [1, 2, 3, 4, 5, 6, 7]
````

**(k) Write a program to find the mean, median and mode of a list of 10 numbers.**

````python
import statistics

lst = [10, 20, 20, 30, 40, 50, 50, 50, 60, 70]
mean = statistics.mean(lst)
median = statistics.median(lst)
mode = statistics.mode(lst)
print("Mean:", mean)       # 40.0
print("Median:", median)   # 40.0
print("Mode:", mode)       # 50
````
