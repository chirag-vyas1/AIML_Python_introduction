## <a id="sub1-1"><a/>**`1.Repetition Control Instruction`**

- It helps us repeat a set of statements in a program. There are two types of repetition control instructions:  
  (a) `while`  
  (b) `for`  

- Unlike many other languages, there is no `do-while` loop in Python.

---

### **`while`** Loop

- `while` is used to repeatedly execute instructions as long as the condition is true.  
  It has two forms:

```python
while condition:
    statement1
    statement2
```
   
```python
while condition:
    statement1
    statement2
else:
    statement3
    statement4
```

- else block is optional. If present, it is executed when the condition fails.
- If the while loop is terminated abruptly using a break statement, then the else block is not executed.

### **`for Loop`**
- for is used to iterate over elements of a sequence such as string, tuple, or list.
It has two forms:

``` python

for var in list:
    statement1
    statement2
```
```python
for var in list:
    statement1
    statement2
else:
    statement3
    statement4
```
- During each iteration, var is assigned the next value from the list.
- In place of a list, a string, tuple, set or dictionary can also be used.
- else block is optional. If present, it is executed if the loop is not terminated abruptly using break.

## <a id="sub1-2"><a/>**`2. Usage of while loop`**

A `while` loop can be used in the following three situations:
- Repeat a set of statements till a condition remains True.
- Repeat a set of statements a finite number of times.
- Iterate through a string, list, and tuple.

---

**Example 1: Repeat until a condition is met (unknown number of times)**
- When we use `while` loop to repeat a set of statements till a condition remains True, it means that when we do not know beforehand how many times the statements are to be executed.

```python
num = int(input('Enter a number: '))
while num != 5:
    print(num, num * num)
    num = int(input('Enter a number: '))
```
- The loop terminates when 5 is entered as input.

**Example 2: Repeat for a fixed number of times**

- We can use a while loop to repeat a set of statements a finite number of times.

``` python
count = 0
while count < 10:
    print(count, count * count, count * count * count)
    count += 1
```

**Example 3: Iterating with an index over string, list, and tuple**
- A while loop can also be used to iterate through a string, a list, or a tuple using an index value.

``` python
s = 'Mumbai'
lst = ['desert', 'dessert', 'to', 'too', 'lose', 'loose']
tpl = (10, 20, 30, -20, -10)
i = 0
while i < len(lst):
    print(i, s[i], lst[i], tpl[i])
    
```
- Since items in a set or a dictionary cannot be accessed using an index value, it is better to use a for loop to access their elements.
- Of the three usages of while loop shown above, the most popular is the first usage  repeat statements an unknown number of times.
- The other two situations are usually handled using a for loop.i += 1

## <a id="sub1-3"><a/>**`3. Usage of for loop`**
A for loop can be used in the following two situations:
- Repeat a set of statements a finite number of times.
- Iterate through a string, list, tuple, set or dictionary.
- To repeat a set of statements a finite number of times, a built-in function `range()` is used.

- **`range()` Function`**

- `range()` generates a sequence of integers.

```python
range(10)        # generates numbers from 0 to 9
range(10, 20)    # generates numbers from 10 to 19
range(10, 20, 2) # generates numbers from 10 to 19 in steps of 2
range(20, 10, -3)# generates numbers from 20 to 9 in steps of -3
```
note : that **range()** can not generate a sequence of float
- In general:
range(start, stop, step)

- Produces a sequence of integers from start (inclusive) to stop (exclusive) by step.
- The list of numbers generated using range() can be iterated through using a for loop.

```python
for i in range(1, 10, 2):
    print(i, i * i, i * i * i)```
```

### **`Iterating with for loop`**
- A for loop is very popularly used to iterate through a string, list, tuple, set, or dictionary, as shown below:

```python 
for char in 'Leopard':
    print(char)

for animal in ['Cat', 'Dog', 'Tiger', 'Lion', 'Leopard']:
    print(animal)

for flower in ['Rose', 'Lily', 'Jasmine']:
    print(flower)

for num in (10, 20, 30, -10, -25):
    print(num)

for key in {'A101': 'Rajesh', 'A111': 'Sunil', 'A112': 'Rakesh'}:
    print(key)
```
- In the first for loop, in each iteration the loop variable char is assigned the next value from the string.
- Similarly, in the second, third, and fourth for loops, in each iteration the loop variable animal / flower / num is assigned the next value from the list / tuple / set.

- Note that in the last `for` loop we are printing only the **keys** in the dictionary.Printing values or printing both keys and values is covered in Chapter 11.

- If while iterating through a collection using a `for` loop we also wish to get an index of the item, we should use the built-in `enumerate()` function as shown below:

```python
lst = ['desert', 'dessert', 'to', 'too', 'lose', 'loose']
for i, ele in enumerate(lst):
    print(i, ele)

## <a id="sub1-4"><a/>**`4. break and continue`**
- break and continue statements can be used with while and for loops.
- break statement terminates the loop without executing the else block.
- continue statement skips the rest of the statements in the block and continues with the next iteration of the loop.

## <a id="sub1-5"><a/>**`5. Else Block of a Loop`**

- else block of a while loop should be used in situations where you wish to execute some statements if the loop is terminated normally and not if it is terminated abruptly.
- Such a situation arises if we are to determine whether a number is prime or not.

```python

i = 2
while i <= num - 1:
    if num % i == 0:
        print(num, 'is not a prime number')

## **`5B else with for Loop - Early Termination Example`**

- In the following example, the `else` block will **not** execute because the list contains `3`, a **non-multiple of 10**. 
- On encountering it, we **terminate** the loop using `break`.

```python
for ele in [10, 20, 30, 3, 40, 50]:
    if ele % 10 != 0:
        print(ele, 'is not a multiple of 10')
        break
else:
    print('All numbers in list are multiples of 10')
```
- The else block only executes if the loop is not terminated using break.

## **`Problem 6.1`**
Write a program that receives 3 sets of values of `p`, `n` and `r` and calculates simple interest for each.
## **`Program`**


````python
i = 1
while i <= 3 :
    p = float(input('Enter value of p: '))
    n = int(input('Enter value of n: '))
    r = float(input('Enter value of r: '))
    si = p * n * r / 100
    print('Simple interest = Rs.' + str(si))
    i = i + 1
````

## **`Problem 6.2`**
Write a program that prints numbers from 1 to 10 using an infinite loop. All numbers should get printed in the same line.
## **`Program`**

````python
i = 1
while 1 :
    print(i,end=' ')
    i += 1
    if i > 10 :
        break
````

## **`Tips`**
- **while 1** creates an infinite loop, as 1 is non-zero, hence true. Replacing 1 with any non-zero number will create an infinite loop.
- Another way of creating an infinite loop is **while True**.
- **end = ' '** in **print()** prints a space after printing i in each iteration. Default value of **end** is newline ('\n').

# **`Problem 6.3`**
Write a program that prints all unique combinations of 1, 2 and 3.
## **`Program`**


````python
i = 1
while i <= 3:
    j = 1
    while j <= 3:
        k = 1
        while k <= 3:
            if i == j or j == k or k == i:
                k += 1
                continue
            else:
                print(i, j, k)
                k += 1
        j += 1
    i += 1
````

## **`[A] Answer the following questions:`**

### `(a) When does the else block of a while loop go to work?`
The else block of a while loop executes when the loop condition becomes false (i.e., when the loop terminates normally, not due to a break statement).

### `(b) Can range() function be used to generate numbers from 0.1 to 1.0 in steps of 0.1?`
No, the range() function only works with integers. It cannot generate floating-point numbers. To generate numbers from 0.1 to 1.0 in steps of 0.1, you would need to use a different approach, such as:
```python
i = 0.1
while i <= 1.0:
    print(round(i, 1))
    i += 0.1
```
### `(c) Can a while loop be nested within a for loop and vice versa?`
Yes, both types of loops can be nested within each other. You can have a while loop inside a for loop, and a for loop inside a while loop.

### `(d) Can a while/for loop be used in an if/else and vice versa?`
Yes, loops can be used inside if/else statements, and if/else statements can be used inside loops.

### `(e) Can a do-while loop be used to repeat a set of statements?`
Python does not have a built-in do-while loop like some other languages. However, you can emulate a do-while loop using a while loop with a condition that's always true and a break statement.

### `(h) Why should break and continue be always used embedded in a while or for loop?`
Break and continue statements are designed to control the flow of loops. Using them outside of a loop context would cause a syntax error because they have no meaning in other contexts.
